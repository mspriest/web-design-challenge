```python
#Import Dependencies
import pandas as pd
from IPython.display import HTML
```

```python
#Read in CSV File & create dataframe
df = pd.read_csv('weather_data.csv')
df.head()
```

```python
#Drop Unnamed:0

df = df.drop(columns=["Unnamed: 0"], axis=1)

print(df)
```

```python
#Convert dataframe to HTML
result=df.to_html()
print(result)
```

```python
#Write HTML script into HTML file
html=df.to_html()

text_file = open("htmldata.html", "w")
text_file.write(html)
text_file.close()
```
